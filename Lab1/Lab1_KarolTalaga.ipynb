# Zadanie 3

Definicja funkcji

```python
def fun(x):
    return x**2+5
```

Zdefiniowanie przedziałów

```python
r1 = np.linspace(-1,1)
r2 = np.linspace(-6,6)
r3 = np.linspace(0,5)
```

Wyrysowanie funkcji na przedziałach, dodanie etykiet osi, legendy i tytułów

```python
_, ax = plt.subplots(1,3, figsize = (15,5))
ax[0].set_title("Przedział 1")
ax[0].plot(r1,fun(r1),'b')
ax[0].set(xlabel='x', ylabel='f(x)', title='f(x) na przedziale [-1,1]')
ax[0].legend(['$f(x)=x^2 + 5$'])
ax[0].grid()

ax[1].set_title("Przedział 2")
ax[1].plot(r2,fun(r2),'g')
ax[1].set(xlabel='x', ylabel='f(x)', title='f(x) na przedziale [-6,6]')
ax[1].legend(['$f(x)=x^2 + 5$'])
ax[1].grid()

ax[2].set_title("Przedział 3")
ax[2].plot(r3,fun(r3),'r')
ax[2].set(xlabel='x', ylabel='f(x)', title='f(x) na przedziale [0,5]')
ax[2].legend(['$f(x)=x^2 + 5$'])
ax[2].grid()

plt.show()
```
